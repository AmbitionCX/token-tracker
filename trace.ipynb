get trace

```python
from dotenv import load_dotenv
import os
import requests
import json
load_dotenv()  # 自动读取当前目录下的 .env 文件

RPC_URL = os.getenv("GETH_API")
print("读取到的 RPC_URL:", RPC_URL)
# 请求部分
def rpc_call(method, params):
    payload = {
        "jsonrpc": "2.0",
        "method": method,
        "params": params,
        "id": 1
    }
    resp = requests.post(RPC_URL, json=payload)
    resp.raise_for_status()
    return resp.json()["result"]
```

```python
# 2. 获取最新区块号
latest_hex = rpc_call("eth_blockNumber", [])
latest_block = int(latest_hex, 16)
print(f"最新区块号: {latest_block}")
```

```python
# 3. 定义函数获取区块信息
def get_block_by_number(block_number):
    return rpc_call("eth_getBlockByNumber", [hex(block_number), True])

def trace_transaction(tx_hash):
    return rpc_call("debug_traceTransaction", [tx_hash, {"tracer": "callTracer"}])
```

```python
# 4. 抓取最近 100 个区块的交易哈希
start_block = latest_block - 1
blocks_data = []

for block_num in range(start_block, latest_block + 1):
    block = get_block_by_number(block_num)
    blocks_data.append(block)
    print(f"获取到区块 {block_num} - 包含 {len(block['transactions'])} 笔交易")
```

```python
# 5. 对每笔交易跑 trace
traces = []

for block in blocks_data:
    for tx in block["transactions"]:
        tx_hash = tx["hash"]
        try:
            trace_result = trace_transaction(tx_hash)
            traces.append({
                "blockNumber": int(block["number"], 16),
                "txHash": tx_hash,
                "trace": trace_result
            })
        except Exception as e:
            print(f"Trace {tx_hash} 失败: {e}")
```

```python
# 6. 保存 trace 结果到本地文件
with open("Results/recent_100_blocks_traces.json", "w", encoding="utf-8") as f:
    json.dump(traces, f, indent=2)

print(f"已保存 {len(traces)} 笔交易 trace 到 recent_100_blocks_traces.json")
```
